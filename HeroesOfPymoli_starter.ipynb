### Heroes Of Pymoli Data Analysis
* Of the 1163 active players, the vast majority are male (84%). There also exists, a smaller, but notable proportion of female players (14%).

* Our peak age demographic falls between 20-24 (44.8%) with secondary groups falling between 15-19 (18.60%) and 25-29 (13.4%).

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
total_calc = len(purchase_data["SN"].unique())
total_number = [{"Total Player": total_calc}]
total_player = pd.DataFrame(total_number)
total_player
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
unique_items = len(purchase_data["Item ID"].unique())
average_price = round(purchase_data["Price"].mean(),2)
purchase_numbers = purchase_data["Purchase ID"].count()
total_revenue = purchase_data["Price"].sum()
```

```python
Summary = [{"Number of Unique Items": unique_items,
            "Average Price": average_price,
            "Number of Purchases": purchase_numbers,
            "Total Revenue": total_revenue            
           }]


df_summary = pd.DataFrame(Summary)
df_summary["Average Price"] = df_summary["Average Price"].map("${:}".format)
df_summary["Total Revenue"] = df_summary["Total Revenue"].map("${:}".format)
df_summary[["Number of Unique Items","Average Price","Number of Purchases","Total Revenue"]]
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
reduced_SN = purchase_data.drop_duplicates("SN")
gender_count = reduced_SN["Gender"].value_counts()
pd_gender_count = pd.DataFrame(gender_count)
pd_gender_count = pd_gender_count.rename(columns={"Gender":"Total Count"})
pd_gender_count["Percentage of Players"] = round((pd_gender_count["Total Count"]/576)*100,2)
pd_gender_count
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
grouped_gender = purchase_data.groupby(["Gender"])
purchase_ID = grouped_gender.count()
purchase_ID = purchase_ID[["Purchase ID"]]
avg_price = grouped_gender.mean()
avg_price = round(avg_price[["Price"]],2)
total_price = grouped_gender.sum()
total_price = round(total_price[["Price"]],2)

combined_1 = pd.merge(purchase_ID, avg_price, on="Gender")
combined_2 = pd.merge(combined_1, total_price, on="Gender")
combined_2 = combined_2.rename(columns={"Purchase ID":"Purchase Count","Price_x":"Average Purchase Price","Price_y":"Total Purchase Value"})
combined_2["Avg Total Purchase per Person"] = round(combined_2["Total Purchase Value"]/pd_gender_count["Total Count"],2)
combined_2["Average Purchase Price"] = combined_2["Average Purchase Price"].map("${:}".format)
combined_2["Total Purchase Value"] = combined_2["Total Purchase Value"].map("${:}".format)
combined_2["Avg Total Purchase per Person"] = combined_2["Avg Total Purchase per Person"].map("${:}".format)
combined_2
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
print(purchase_data["Age"].max())
```

```python
bins = [0,9,14,19,24,29,34,39,45]

group_labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34",
                "35-39", "40+"]

age_group = pd.cut(reduced_SN["Age"], bins, labels=group_labels)
pd_age_group = pd.DataFrame(age_group)
age_summary = pd_age_group["Age"].value_counts()
pd_age_summary = pd.DataFrame(age_summary)
pd_age_summary = pd_age_summary.sort_index(ascending=True)
pd_age_summary = pd_age_summary.rename(columns={"Age":"Total Count"})
pd_age_summary["Percentage of Players"] = round((pd_age_summary["Total Count"]/576)*100,2)
pd_age_summary
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase_data["Age Group"] = pd.cut(purchase_data["Age"], bins, labels=group_labels)
purchase_analysis = purchase_data.groupby("Age Group")
avg_purchase_price = purchase_analysis.mean()
avg_purchase_price = round(avg_purchase_price[["Price"]],2)
purchase_count = purchase_analysis.count()
purchase_count = purchase_count[["Purchase ID"]]
total_purchase = purchase_analysis.sum()
total_purchase = round(total_purchase [["Price"]],2)

age_combined = pd.merge(purchase_count, avg_purchase_price, on="Age Group")
age_combined2 = pd.merge(age_combined, total_purchase, on="Age Group")
age_combined2 = age_combined2.rename(columns={"Purchase ID":"Purchase Count","Price_x":"Average Purchase Price","Price_y":"Total Purchase Value"})
age_combined2["Avg Total Purchase per Person"] = round(age_combined2["Total Purchase Value"]/pd_age_summary["Total Count"],2)
age_combined2["Average Purchase Price"] = age_combined2["Average Purchase Price"].map("${:}".format)
age_combined2["Total Purchase Value"] = age_combined2["Total Purchase Value"].map("${:}".format)
age_combined2["Avg Total Purchase per Person"] = age_combined2["Avg Total Purchase per Person"].map("${:}".format)
age_combined2
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
top_spender = purchase_data.groupby("SN")
purchase_count = top_spender.count()
purchase_count = purchase_count[["Purchase ID"]]
purchase_price = top_spender.mean()
purchase_price = round(purchase_price[["Price"]],2)
purchase_total = top_spender.sum()
purchase_total = round(purchase_total[["Price"]],2)

spender_combined = pd.merge(purchase_count, purchase_price, on="SN")
spender_combined2 = pd.merge(spender_combined, purchase_total, on="SN")
spender_combined2 = spender_combined2.rename(columns={"Purchase ID":"Purchase Count","Price_x":"Average Purchase Price","Price_y":"Total Purchase Value"})
spender_combined2 = spender_combined2.sort_values("Total Purchase Value", ascending = False)
spender_combined2["Average Purchase Price"] = spender_combined2["Average Purchase Price"].map("${:}".format)
spender_combined2["Total Purchase Value"] = spender_combined2["Total Purchase Value"].map("${:}".format)
spender_combined2.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
most_items = purchase_data.groupby(["Item ID","Item Name"])
count = most_items.count()
count = count[["Purchase ID"]]
avg = most_items.mean()
avg = round(avg[["Price"]],2)
total = most_items.sum()
total = round(total[["Price"]],2)

most_combined = pd.merge(count, avg, on=["Item ID","Item Name"])
most_combined2 = pd.merge(most_combined, total, on=["Item ID","Item Name"])
most_combined2 = most_combined2.rename(columns={"Purchase ID":"Purchase Count","Price_x":"Item Price","Price_y":"Total Purchase Value"})
most_combined2 = most_combined2.sort_values("Purchase Count", ascending = False)
most_combined2["Item Price"] = most_combined2["Item Price"].map("${:}".format)
most_combined2["Total Purchase Value"] = most_combined2["Total Purchase Value"].map("${:}".format)
most_combined2.head()
```
